```python
import requests

def text_to_speech(filename, text):
    server_url = "http://54.179.185.193:10010"
    endpoint = "/tts/"
    params = {"text": text}

    response = requests.get(server_url + endpoint, params=params)

    if response.status_code == 200:
        filename = filename  # 저장할 파일 이름
        with open(filename, "wb") as f:
            f.write(response.content)
        print(f"음성 파일이 '{filename}'으로 저장되었습니다.")
    else:
        print(f"오류 발생: {response.status_code}")

text_to_speech("tts_output.wav", "Did he write stories when he was younger? I think he used to write novels.")
```

```python
def test_stt_endpoint(filename):
    server_url = "http://127.0.0.1:10010"  # FastAPI 서버 주소 (필요에 따라 수정)
    endpoint = "/stt/"
    audio_file_path = filename  # 테스트할 음성 파일 경로

    with open(audio_file_path, "rb") as f:
        files = {"file": (audio_file_path, f)}
        response = requests.post(server_url + endpoint, files=files)

    if response.status_code == 200:
        result = response.json()
        transcription = result.get("transcription")
        if transcription:
            return transcription
        else:
            return result.get("error")
    else:
        print(f"오류 발생: {response.status_code}")

# 테스트 실행
test_stt_endpoint('tts_output.wav')
```
